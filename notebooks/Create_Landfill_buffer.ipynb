```python
import geopandas as gp
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
lm = '../data/output/landfills_master.csv'
```

```python
lm = pd.read_csv(lm)
```

```python
def geom(x):
    return Point(x['lng'], x['lat'])
```

```python
lm['geometry'] = lm.apply(geom, axis=1)
```

```python
lmm = gp.GeoDataFrame(lm)
```

```python
lmm.crs = {'init':'epsg:4326'}
```

```python
fname = '../data/output/active_msw_landfills.shp'
```

```python
df = gp.read_file(fname)
```

- Latitude: 1 deg = 110.574 km
- Longitude: 1 deg = 111.320*cos(latitude) km

```python
def latlng_to_km(km):
    '''
    Parameters
    ----------
    km: float,
        Distance in km to be converted latlng
    '''
    return km * (1./110.574)
```

```python
# get bounds for each geometry
bounds = df['geometry'].bounds

# calc distance for buffer
radius = latlng_to_km(100)

# adjust bounds
bounds.minx = bounds.minx - radius
bounds.miny = bounds.miny - radius
bounds.maxx = bounds.maxx + radius
bounds.maxy = bounds.maxy + radius
```

```python
pd.concat([lmm,bounds], axis=1)
```
